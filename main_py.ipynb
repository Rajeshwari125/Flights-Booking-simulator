```
import sqlite3

# Create DB
conn = sqlite3.connect("flights.db")
cursor = conn.cursor()

# Create flights table
cursor.execute("""
CREATE TABLE IF NOT EXISTS flights (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    origin TEXT NOT NULL,
    destination TEXT NOT NULL,
    departure_time TEXT NOT NULL,
    arrival_time TEXT NOT NULL,
    base_price INTEGER NOT NULL,
    available_seats INTEGER NOT NULL
)
""")

conn.commit()
conn.close()

print("Database & table created successfully!")
```

```
import sqlite3

flights_data = [
    ("Chennai", "Mumbai", "2025-11-25 08:00", "2025-11-25 10:00", 3500, 60),
    ("Chennai", "Delhi", "2025-11-25 09:00", "2025-11-25 12:10", 5000, 50),
    ("Bangalore", "Chennai", "2025-11-26 07:30", "2025-11-26 08:40", 2500, 40),
    ("Hyderabad", "Chennai", "2025-11-27 05:00", "2025-11-27 07:00", 2800, 70),
]

conn = sqlite3.connect("flights.db")
cursor = conn.cursor()

cursor.executemany("""
INSERT INTO flights (origin, destination, departure_time, arrival_time, base_price, available_seats)
VALUES (?, ?, ?, ?, ?, ?)
""", flights_data)

conn.commit()
conn.close()

print("Sample flight data inserted!")
```

```
from fastapi import FastAPI
import sqlite3

app = FastAPI()

# Function to fetch data
def get_flights(query, params=()):
    conn = sqlite3.connect("flights.db")
    cursor = conn.cursor()
    cursor.execute(query, params)
    rows = cursor.fetchall()
    conn.close()
    return rows

@app.get("/flights")
def all_flights():
    flights = get_flights("SELECT * FROM flights")
    return {"flights": flights}

@app.get("/search")
def search_flights(origin: str, destination: str):
    flights = get_flights(
        "SELECT * FROM flights WHERE origin=? AND destination=?",
        (origin, destination)
    )
    return {"results": flights}
```
